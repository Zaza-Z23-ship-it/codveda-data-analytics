```python

import pandas as pd
data ={
'CustomerID' : [1, 2, 3, 4, 2, 5, 6,],
'Name' : ['Alice', 'Bob', 'Carol', 'David', 'Bob', 'Eve', 'Frank'],
'Age' : [25, None, 30, 22, None, 28, 35],
'City' : ['Lagos', 'Abuja', 'lagos', 'Kano', 'Abuja', 'Lagos', None],
'JoinDate' : ['2024-01-15', '2024/02/20', '15-03-2024', '2024-04-10', '2024/02/20', '2024-05-05', '2024-06-18'],
'Purchase' : [5000, 3000, 4500, 2000, 3000, 6000, 1500]
}
df = pd.DataFrame(data)
df.to_csv('messy_customers.csv', index=False)
print("Dataset created!")
print(df)    
```

```python
average_age = df['Age'].mean()
print(f"Average age: {average_age:.1f}")
df['Age'] = df['Age'].fillna(average_age)
print(" Missing ages filled with average age")
print(df['Age'])
```

```python
df = df.dropna(subset=['City'])
print("Rows with missing City removed")
print(f"Remaining rows: {len(df)}")
```

```python
print(f"Duplicates found: {df.duplicated().sum()}")
df = df.drop_duplicates()
print(f"Duplicates removed. Remaining rows: {len(df)}")
```

```python
df['City'] = df ['City'].str.strip()       # Remove extra spaces
df['City'] = df ['City'].str.title()       # Converts to Title Case
# str.title() turns "lagos" → "Lagos", "ABUJA" → "Abuja"

print("City names standardized:")
print(df['City'].unique())  # .unique() shows all unique values in the column
```

```python
df['JoinDate'] = pd.to_datetime(df['JoinDate'], dayfirst=True, errors='coerce')
# pd.to_datetime() is smart — it tries to understand different date formats
# dayfirst=True → helps with dates like '15-03-2024' (day comes first)
# errors='coerce' → if it can't understand a date, put NaT (Not a Time) instead of crashing

# Now format all dates the same way: YYYY-MM-DD
df['JoinDate'] = df['JoinDate'].dt.strftime('%Y-%m-%d')
print(" Dates standardized:")
print(df['JoinDate'])
```

```python
print("=== CLEAN DATA (After Cleaning) ===")
print(df)

# Reset the index (row numbers) since we removed some rows
df = df.reset_index(drop=True)

# Save the cleaned dataset as a new CSV file
df.to_csv('clean_customers.csv', index=False)
print(" Clean dataset saved as 'clean_customers.csv'")
# Final summary report
print("=== CLEANING SUMMARY ===")
print(f"Remaining rows: {len(df)}")
print(f"Remaining columns: {len(df.columns)}")
print(f"Missing values remaining: {df.isnull().sum().sum()}")
print(f"Duplicate rows remaining: {df.duplicated().sum()}")
print("\nData Types:")
print(df.dtypes)
```

```python
import pandas as pd

# ── 1. LOAD ──────────────────────────────────────────────
df = pd.read_csv('messy_customers.csv')
print("Raw data loaded:", df.shape)

# ── 2. MISSING VALUES ────────────────────────────────────
df['Age'] = df['Age'].fillna(df['Age'].mean())   # Fill age with mean
df = df.dropna(subset=['City'])                   # Drop rows with missing city

# ── 3. DUPLICATES ────────────────────────────────────────
df = df.drop_duplicates()

# ── 4. STANDARDIZE ───────────────────────────────────────
df['City'] = df['City'].str.strip().str.title()  # Fix capitalization
df['JoinDate'] = pd.to_datetime(df['JoinDate'], dayfirst=True, errors='coerce')
df['JoinDate'] = df['JoinDate'].dt.strftime('%Y-%m-%d')  # Uniform date format

# ── 5. RESET & SAVE ──────────────────────────────────────
df = df.reset_index(drop=True)
df.to_csv('clean_customers.csv', index=False)
print("Clean data saved!")
print(df)
```
