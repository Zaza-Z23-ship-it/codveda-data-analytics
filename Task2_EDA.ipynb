```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('clean_customers.csv')
print("Data loaded successfully!")
print(df)
```

```python
print("DATASET SHAPE")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\n COLUMN NAMES & DATA TYPES ")
print(df.dtypes)
print("\n FIRST 5 ROWS")
print(df.head())
print("\n LAST 5 ROWS")
print(df.tail())
```

```python
print("SUMMARY STATISTICS")
print(df.describe())
print("\n What these mean:")
print("Count  = how many values exist")
print("mean  = average value")
print("std  = standard deviation")
print("min  = smallest value")
print("25%  = 25% of values are below this")
print("50%  = middle value (median)")
print("75%  = 75% of values are below this")
print("max  = largest value")
```

```python
print("PURCHASE AMOUNT STATISTICS ")
print(f"Mean   (Average): {df['Purchase'].mean():.2f}")
print(f"Median (Middle):  {df['Purchase'].median():.2f}")
print(f"Mode   (Most common): {df['Purchase'].mode()[0]:.2f}")
print(f"Std Dev (Spread): {df['Purchase'].std():.2f}")
print(f"Min:  {df['Purchase'].min():.2f}")
print(f"Max:  {df['Purchase'].max():.2f}")

print("\nAGE STATISTICS ")
print(f"Mean Age:   {df['Age'].mean():.1f}")
print(f"Median Age: {df['Age'].median():.1f}")
print(f"Youngest:   {df['Age'].min():.0f}")
print(f"Oldest:     {df['Age'].max():.0f}")
print(" PURCHASE AMOUNT STATISTICS ")
print(f"Mean   (Average): {df['Purchase'].mean():.2f}")
print(f"Median (Middle):  {df['Purchase'].median():.2f}")
print(f"Mode   (Most common): {df['Purchase'].mode()[0]:.2f}")
print(f"Std Dev (Spread): {df['Purchase'].std():.2f}")
print(f"Min:  {df['Purchase'].min():.2f}")
print(f"Max:  {df['Purchase'].max():.2f}")

print("\n AGE STATISTICS")
print(f"Mean Age:   {df['Age'].mean():.1f}")
print(f"Median Age: {df['Age'].median():.1f}")
print(f"Youngest:   {df['Age'].min():.0f}")
print(f"Oldest:     {df['Age'].max():.0f}")
```

```python
plt.figure(figsize=(8, 5))   # Set the size of the chart

plt.hist(df['Purchase'], 
         bins=5,              # Divide data into 5 groups
         color='steelblue', 
         edgecolor='black')

plt.title('Distribution of Purchase Amounts')   # Chart title
plt.xlabel('Purchase Amount ')               # X-axis label
plt.ylabel('Number of Customers')               # Y-axis label
plt.grid(axis='y', alpha=0.5)                   # Light grid lines

plt.tight_layout()
plt.savefig('histogram_purchase.png')           # Save as image
plt.show()
print(" Histogram saved")
```

```python
plt.figure(figsize=(8, 5))

sns.boxplot(data=df[['Age', 'Purchase']], palette='Set2')

plt.title('Boxplot of Age and Purchase Amount')
plt.ylabel('Value')
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.savefig('boxplot.png')
plt.show()
print("Boxplot saved!")
```

```python
print(" CORRELATION BETWEEN NUMERICAL COLUMNS ")
correlation = df[['Age', 'Purchase']].corr()
print(correlation)

# Visualize correlation as a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, 
            annot=True,        # Show numbers on the heatmap
            cmap='coolwarm',   # Color scheme
            fmt='.2f',         # Round to 2 decimal places
            linewidths=0.5)

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()
print("Correlation heatmap saved")
```
